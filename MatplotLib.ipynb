#  MatplotLib Tutorial 
### MatplotLib is a plotting library for the Python programming langugage and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter,wxPython, QT or GTK+.

```python
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
import numpy as np
```

```python
## simple Example

x= np.arange(0,10)
y= np.arange(11,21)
```

```python
# plotting using matplotlib

## plt scatter

plt.scatter(x,y,c='y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
y=x*x
```

```python
## plot plot 

plt.plot(x,y,'rp',linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('x-axis')
plt.ylabel('Y-axis')
plt.title('2d Diagraam')
plt.show()
```

```python
## creating Subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'yo')
```

```python
x = np.arange(1,11)
y= 3*x + 5
plt.title("Matplotlib Demo")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y)
```

```python
# compute the x and y  coordinate for points on a sine curve
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title("sine wave form")
plt.plot(x,y)
plt.show()
```

```python
# Subplot
## sine and cosine curve

x= np.arange(0, 5 * np.pi,0.1)
y_sine = np.sin(x)
y_cos = np.cos(x)
# set the first subplot
plt.subplot(2,1,1)

#make the first plot
plt.plot(x,y_sine,'r--')
plt.title("Sine curve")

#set the second subplot
plt.subplot(2,1,2)
#make the second plot
plt.plot(x,y_cos,'g--')
#plt.title("Cosine curve")
```

```python
# bar plot

x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bar Graph")
```

# Histograms

```python
a = np.array([22,87,5,42,11,55,99,56,54,25,74,85,12])
plt.hist(a)
plt.title("Histogram")
plt.show()
```

# Box Plot using matplotlib

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True)
plt.show()
```

# Pichat

```python
# data to plot

label = 'Python','c++','Ruby',"java"
sizes = [215,130,245,210]
colors = ['gold','green','yellow','lightblue']
explode = (0,0.1,0.2,0) # explode 1st slice

#plot 
plt.pie(sizes, explode=explode, labels=label,colors=colors,
        autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.title("Piechart")
plt.show()
```
