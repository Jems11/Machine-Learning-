### Practise problem on IRIS Dataset

```python
import seaborn as sns
```

# JoinPlot

# Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

# Pair Plot

 A Pair plot is laso known as scatterplot, in which one variable in the same data row is matched with another variable's value 

```python
sns.pairplot(df)
```

```python
df['time'].value_counts()
```

```python
sns.pairplot(df,hue='time')
```

# Dist Plot

Dist Plot helps us to check the distribution of the column feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

# Categorical Plots

Seaborn also helps us in doing the analysis on categorical Data points. In this section we will discuss abut 
- boxplot
- violinplot 
- countplot
- bar plot

# 1. Count plot

```python
## count plot
sns.countplot('day',data=df)
```

```python
df['time'].value_counts()
```

```python
sns.countplot(y="sex",data=df)
```

# 2. Bar PLot

```python
sns.barplot(x="total_bill",y="smoker",data=df)
```

```python
df.head()
```

# 3. Boxplot

A box plot and whisker plot (sometimes called a boxplot) is a graph that presents information from a five number summary.

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df,palette='Set2') 
```

```python
sns.boxplot(data=df,orient='v')
```

```python
## catoegorize my data based on some other categories

sns.boxplot("total_bill","day",hue="sex",data=df)
```

# 4. Violin Plot 

Violin plot helps us to see both the distribution of data in terms of kernal density estimation and the box plot.

```python
sns.violinplot('total_bill','smoke',data=df)
```
